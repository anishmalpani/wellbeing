```python
imf = pd.read_csv ('imf_oct_2019.csv')
imf.head()
```

```python
s = imf[['WEO Country Code', 'ISO', 'WEO Subject Code', 'Country']]
s.head()
```

```python
imf_melt = pd.melt(imf, id_vars=['WEO Country Code', 'ISO', 'WEO Subject Code', 'Country', 'Subject Descriptor'])
imf_melt.head()
```

```python
imf_ah = pd.pivot_table (imf_melt, index =['WEO Country Code', 'ISO', 'Country', 'variable'], columns = 'Subject Descriptor', values = 'value', aggfunc=np.sum) 
```

```python
imf_ah.reset_index(inplace=True)
```

```python
imf_ah.head()
```

```python
imf_ah.drop (columns = ['level_4'], inplace=True)
imf_ah.head()
```

```python
imf_ah.rename (columns = {
    'variable': 'date',
    'ISO': 'country_code'
}, inplace=True)
imf_ah.head()
```

```python
imf_ah.to_csv ('imf_2019_output.csv', index=False)
```
